```python
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X, y = mnist.data, mnist.target
```

```python
X /= 255.
X = X[:5000]
y = y[:5000]
y = np.array([int(e) for e in y])
```

```python
class DisjointSet:
    def __init__(self, n):
        self.s = [-1]*n

    def find(self, a):
        if self.s[a] < 0:
            return a
        parent = self.find(self.s[a])
        self.s[a] = parent
        return parent

    def sameset(self, a, b):
        return self.find(a) == self.find(b)

    def union(self, a, b):
        if self.sameset(a, b):
            return
        a = self.find(a)
        b = self.find(b)
        if -self.s[a] > -self.s[b]:
            self.s[a] += self.s[b]
            self.s[b] = a
        else:
            self.s[b] += self.s[a]
            self.s[a] = b
```

```python
def kruskal(G, k=2):
    n = len(G)
    ds = DisjointSet(n)
    edges = [(G[u, v], u, v) for u in range(n) for v in range(n)]
    edges.sort()
    #T = []
    links = 0
    for _, u, v in edges:
        if not ds.sameset(u, v):
            #T.append((u, v))
            ds.union(u, v)
            links += 1
        if links == n - k: break

    y_ = np.zeros((n), dtype=int)
    unique = set()
    for u in range(n):
        unique.add(ds.find(u))
        y_[u] = ds.find(u)

    for i, u in enumerate(unique):
        y_[y_ == u] = i

    return y_
```

```python
def eucli(p1, p2):
    return sum((p1 - p2)**2)**0.5

assert eucli(np.array([0, 3]), np.array([4, 0])) == 5.
assert eucli(np.array([0, 0, 0]), np.array([1, 1, 1])) == 3**0.5
```
